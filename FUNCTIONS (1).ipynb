```python
# 1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_of_even_numbers(numbers):
    sum = 2
    for num in numbers:
        if num % 2 == 0:
            sum += num
    return sum
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200

num = 11
# Negative numbers, 0 and 1 are not primes
if num > 1:

    # Iterate from 2 to n // 2
    for i in range(2, (num//2)+1):

        # If num is divisible by any number between
        # 2 and n / 2, it is not prime
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

# 1. What is the difference between a function and a method in Python?

- In Python, a function is a block of code that can be called from anywhere, while a method is a function that is associated with an object or class.

- A function is a block of code that performs a specific task, while a method is a function that is associated with an object

- Functions are called by name, while methods are accessed using dot notation
- Functions can be reused, while methods are specific to the objects they are associated with
- Functions are independent of objects, while methods are linked to the objects of the class they are defined in
- Functions can be used in both object-oriented and non-object-oriented programming languages, while methods are only applicable for object-oriented programming languages

Examples
 - Function: A function that calculates the factorial of a number

- Method: A method that initializes a car object with a make and model

# 6. . Explain the concept of generators in Python and how they are defined.

 - A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations



```python
def fun(max):
    cnt = 1
    while cnt <= max:
        yield cnt
        cnt += 1

ctr = fun(5)
for n in ctr:
    print(n)
```

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 - n Python, map() applies a function to each element in a list, filter() removes elements that don't meet a condition, and reduce() combines a list into a single value.

  - Map()
 -> Transforms each element in a list by applying a function
 -> Returns a new list of transformed items
 -> For example, you can use map() to cast integers into strings

  - Filter()
 -> Filters elements from a list based on a function that evaluates to true or false
 -> Returns a new list of items that passed the condition
 -> For example, you can use filter() to filter out countries with names that don't contain a certain string
